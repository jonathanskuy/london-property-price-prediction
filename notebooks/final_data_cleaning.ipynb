````python
# Examining the distribution of raw historical property prices from the property dataset

plt.figure()
plt.hist(df_property['history_price'], bins=50)
plt.ticklabel_format(style='plain', axis='x')
plt.xlabel('Property Price (£)')
plt.xticks(rotation=45)
plt.ylabel('Frequency')
plt.title('Distribution of Property Prices (Raw)')
plt.show()
````

The raw historical property price data is heavily right-skewed. To solve this and balance the distribution, the prices will be log-transformed.

````python
# Log-transforming raw historical property prices from the property dataset

df_property['log_price'] = np.log(df_property['history_price'])

df_property
````

````python
# Examining the distribution of log-transformed historical property prices from the property dataset

plt.figure()
plt.hist(df_property['log_price'], bins=50)
plt.ticklabel_format(style='plain', axis='x')
plt.xlabel('Log-Transformed Property Price (£)')
plt.xticks(rotation=45)
plt.ylabel('Frequency')
plt.title('Distribution of Property Prices (Log-Transformed)')
plt.show()
````

As seen in the histogram above, log-transforming property prices balanced the distribution from being right-skewed.

````python
# Dropping duplicate rows from the property dataset

df_property = df_property.drop_duplicates()

df_property
````

````python
# Investigating records with identical price, date, and coordinates

df_property[df_property.duplicated(subset=['history_price', 'history_date', 'latitude', 'longitude'], keep=False)].head(10)
````

````python
# Examining property tenure types

df_property['tenure'].unique()
````

For modern properties, there are generally only two types of tenures: Freehold and Leasehold. 'Feudal' leaseholds are outdated, and Shared tenures are rare.

````python
# Filtering out properties with 'Shared' tenure

df_property = df_property[df_property['tenure'] != 'Shared']

df_property
````

After filtering out properties with 'Shared' tenure, only 15 rows were removed which is insignificant compared to the amount of rows the dataset currently has.

````python
# Investigating types of properties with 'Feudal' leasehold tenure

df_property[df_property['tenure'] == 'Feudal']['propertyType'].value_counts()
````

Most of the properties with 'Feudal' leasehold tenure are flats. Therefore, it is safe to assume that these properties are leaseholds.

````python
# Replacing Feudal tenures with Leasehold

df_property['tenure'] = df_property['tenure'].replace('Feudal', 'Leasehold')

df_property['tenure'].value_counts()
````

````python
# Investigating properties with duplicate address, postcode, price, date and coordinates

df_property[df_property.duplicated(subset=['fullAddress', 'postcode', 'history_price', 'history_date', 'latitude', 'longitude'], keep=False)]
````

````python
# Checking for missing values

df_property.isna().sum()
````

There are records with duplicate address, postcode, price, date and coordinates. This usually means that it is a true duplicate transaction, which needs to be removed. Therefore, sorting the dataset by property type, tenure, number of bedrooms/bathrooms/living rooms, and floor area and then removing these duplicates is a safe way to keep only more-'sensible' records.

````python
# Sorting the data by property type, tenure, number of bedrooms/bathrooms/living rooms, and floor area, as well as dropping duplicate records afterwards

df_property = df_property.sort_values(['propertyType', 'tenure', 'bedrooms', 'bathrooms', 'livingRooms', 'floorAreaSqM']).drop_duplicates(subset=['fullAddress', 'history_price', 'history_date', 'latitude', 'longitude'])

df_property
````

````python
# Investigating records with duplicate address, price, date, and coordinates

df_property[df_property.duplicated(subset=['fullAddress', 'history_price', 'history_date', 'latitude', 'longitude'], keep=False)].head(10)
````

After dropping, there are no more duplicate records.

````python
# Function to standardise postcode formats

def standardise_postcodes(df, col):
    df[col] = (df[col].astype(str).str.upper().str.strip().str.replace(r'[^A-Z0-9]', '', regex=True))
    df = df[df[col].notna()]
    df = df[df[col] != '']

    postcode_regex = r'^[A-Z]{1,2}[0-9][0-9A-Z]?\s?[0-9][A-Z]{2}'
    df = df[df[col].str.match(postcode_regex)]

    df[col] = df[col].str[:-3] + ' ' + df[col].str[-3:]

    return df
````

````python
# Standardising postcode formats for the property dataset

df_property = standardise_postcodes(df_property, 'postcode')

df_property
````

````python
# Keeping only relevant columns in the school dataset

df_schools = df_schools[['URN', 'Postcode', 'As at date', 'Phase', 'Overall effectiveness']]

df_schools
````

````python
# Standardising postcode formats in the ONS Postcode Directory dataset

df_onspd = standardise_postcodes(df_onspd, 'postcode')

df_onspd
````

````python
# Standardising postcode formats in the school dataset

df_schools = standardise_postcodes(df_schools, 'Postcode')

df_schools
````

````python
# Merging the ONS Postcode Directory dataset with the school dataset to obtain coordinates

df_schools = df_schools.merge(df_onspd, left_on='Postcode', right_on='postcode', how='left')

df_schools
````

````python
# Reviewing data types in the school dataset

df_schools.dtypes
````

Earlier, it can be observed that there was an inconsistency with the format of the first record compared to the rest in the school dataset. Since this case is only found in the first record, its format will be fixed manually.

````python
# Manually fixing the format of the first record in the school dataset

df_schools['URN'][0] = 100000
df_schools['As at date'][0] = '2024-12-31T00:00:00'
df_schools['Phase'][0] = 'Primary'
df_schools['Overall effectiveness'][0] = 1

df_schools
````

````python
# Dropping the duplicate postcode column from merging

df_schools = df_schools.drop(columns=['postcode'])

df_schools
````

Earlier, the data type for the date each inspection is accurate of is ```object``` where it should be ```datetime``` in order for the dates to be filtered later on. This will be fixed in the next cell.

````python
# Converting the date column in the school dataset into datetime

df_schools['As at date'] = pd.to_datetime(df_schools['As at date'])

df_schools
````

````python
# Filtering dates to before or on 2024-9-30, which is the latest date in the property dataset, to maintain temporal consistency

df_schools = df_schools[df_schools['As at date'] <= '2024-9-30']

df_schools
````

````python
# Sorting the values by date and URN, and filtering duplicate schools by keeping the latest school inspection record for each school

df_schools = df_schools.sort_values(['As at date', 'URN']).groupby('URN', as_index=False).tail(1)

df_schools
````

````python
# Exporting the school dataset into a CSV file

df_schools.to_csv('../data/processed/schools_clean.csv', index=False)
````
