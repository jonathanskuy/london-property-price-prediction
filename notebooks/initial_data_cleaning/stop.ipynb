This notebook contains some of the processes for the Initial Data Cleaning phase, specifically for cleaning the public transport stops dataset. This notebook covers the process of cleaning the public transport stops dataset.

````python
# Importing libraries

import pandas as pd
import os
````

````python
# Listing the directory

print(os.listdir())
````

````python
# Loading the public transport stops dataset

df = pd.read_csv("../../data/raw/Stops.csv")

print("Dataset Shape:", df.shape)
df.head()
````

````python
# Reviewing the public transport stops dataset's info

df.info()
````

````python
# Reviewing columns in the public transport stops dataset

df.columns
````

````python
# Reviewing descriptive statistics for the public transport stops dataset

df.describe()
````

````python
# Standardising column names in the public transport stops dataset into format easier to type

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.columns
````

````python
# Dropping empty rows

df = df.dropna(how='all')

print("After removing empty rows:", df.shape)
````

````python
# Dropping duplicate rows

df = df.drop_duplicates()

print("After removing duplicates:", df.shape)
````

````python
# Filling missing values with "Unknown"

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna("Unknown")
````

Earlier, there are date columns in the public transport stops dataset that are not the ```datetime``` data type. This will be fixed in the next cell.

````python
# Converting date columns into datetime

if 'creationdatetime' in df.columns:
    df['creationdatetime'] = pd.to_datetime(df['creationdatetime'], errors='coerce')

if 'modificationdatetime' in df.columns:
    df['modificationdatetime'] = pd.to_datetime(df['modificationdatetime'], errors='coerce')
````

````python
# Ensuring that columns with numerical values are of numeric data types

for col in ['latitude', 'longitude', 'easting', 'northing']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
````

````python
# Reviewing the shape of the public transport stops dataset

df.shape
````

````python
# Exporting the public transport stops dataset into a CSV file

df.to_csv("../../data/partially_processed/stops.csv", index=False)

print("Cleaned file saved successfully")
````
