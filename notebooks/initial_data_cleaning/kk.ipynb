This notebook contains some of the processes for the Initial Data Cleaning phase, specifically for cleaning the ONS Postcode Directory dataset. This notebook covers the process of cleaning the ONS Postcode Directory dataset.

````python
# Importing libraries

import pandas as pd
````

````python
# Loading the ONS Postcode Directory dataset

df = pd.read_csv("../../data/raw/ONSPD_AUG_2025_UK.csv", low_memory=False)
df.head()
````

````python
# Reviewing columns in the ONS Postcode Directory dataset

df.columns
````

````python
# Keeping only relevant columns in the ONS Postcode Directory dataset

df = pd.read_csv(
    "../../data/raw/ONSPD_AUG_2025_UK.csv",
    usecols=["pcds", "lat", "long"],
    low_memory=False
)

df.head()
````

````python
# Standardising column names in the ONS Postcode Directory dataset into more-readable formats

df = df.rename(columns={"pcds": "postcode", "lat": "latitude", "long": "longitude"})
df.head()
````

````python
# Standardising postcodes in the ONS Postcode Directory dataset

df["postcode"] = df["postcode"].astype(str).str.upper().str.replace(" ", "", regex=False)
df.head()
````

````python
# Ensuring that the coordinates are of numeric data types

df["latitude"] = pd.to_numeric(df["latitude"], errors="coerce")
df["longitude"] = pd.to_numeric(df["longitude"], errors="coerce")
````

````python
# Dropping rows with missing values and ensuring that only postcodes within the UK are included in the ONS Postcode Directory dataset

df = df.dropna(subset=["postcode", "latitude", "longitude"])
df = df[(df["latitude"].between(49, 61)) & (df["longitude"].between(-9, 3))]  # UK approx range
````

````python
# Dropping rows with duplicate postcodes in the ONS Postcode Directory dataset

df = df.drop_duplicates(subset=["postcode"])
````

````python
# Reviewing descriptive statistics of columns in the ONS Postcode Directory dataset

print("Rows:", len(df))
df.describe(include="all")
````

````python
# Exporting the ONS Postcode Directory dataset into a CSV file

df.to_csv("../../data/partially_processed/onspd.csv", index=False)
print("Saved: onspd.csv")
````
